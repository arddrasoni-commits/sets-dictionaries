```python
sounds = {"bass", "drums", "guitar"}
print("bass" in sounds)
print("keys" in sounds)
```

```python
# once a set is created, you cannot replace its items, but you can add new items
# Set items are unchangeable, but you can remove items and add new items.
# Remove and Add items
sounds = {"drums","bass", "guitar"}
sounds.remove("drums")
sounds.add("keys")
print(sounds)
```

```python
# Set operations
S = {"guitar", "harp", "bass", "banjo"}
P = {"guitar", "keys", "bass", "drums"}
print ("Union: ", S|P)
print("Intersection: ", S & P)
print ("Difference: ", S - P)
print ("Symmetric Difference: ", S ^ P)
```

```python
#frozen set
# Addition or removal of elements are not allowed in frozenset
u = frozenset({"guitar", "bass", "drums"})
print(u)
print(type(u))
```
